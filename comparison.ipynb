```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ref=np.genfromtxt('output/near_field2/source_field.txt',delimiter=',')
ref=ref[...,0]+1j*ref[...,1]
ref=ref.reshape(101,-1)
#ref=ref/ref[50,50]
test=np.genfromtxt('output/near_field3/source_field.txt',delimiter=',')
test=test[...,0]+1j*test[...,1];
test=test.reshape(101,-1)
#test=test/test[50,50]
```
