```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_field(folder):
    data=np.genfromtxt(folder +'IF_field.txt',delimiter=',')
    N_if = int(np.sqrt(data.shape[0]))
    ifx = data[...,0]+1j*data[...,1]
    ifx = ifx.reshape(N_if,-1)
    factor =ifx[int(N_if/2),int(N_if/2)]
    ifx = ifx/factor
    ify = data[...,2]+1j*data[...,3]
    ify = ify.reshape(N_if,-1)
    ifz = data[...,4]+1j*data[...,5]
    ifz = ifz.reshape(N_if,-1)

    data=np.genfromtxt(folder +'source_field.txt',delimiter=',')
    N_s = int(np.sqrt(data.shape[0]))
    sx = data[...,0]+1j*data[...,1]
    sx = sx.reshape(N_s,-1)
    factor =sx[int(N_s/2),int(N_s/2)]
    sx =sx/factor
    sy = data[...,2]+1j*data[...,3]
    sy = sy.reshape(N_s,-1)
    sz = data[...,4]+1j*data[...,5]
    sz = sz.reshape(N_s,-1)
    return ifx,ify,ifz,sx,sy,sz
```

```python
outputfolder = 'output/infocus/GPU/296GHz/'
ref_ifx,ref_ify,ref_ifz,ref_sx,ref_sy,ref_sz = read_field(outputfolder+'401_401/')
```

```python
IF_15_x,IF_15_y,IF_15_Z,S_15_x,S_15_y,S_15_z= read_field(outputfolder+'15_15/')
IF_17_x,IF_17_y,IF_17_Z,S_17_x,S_17_y,S_17_z= read_field(outputfolder+'17_17/')
IF_21_x,IF_21_y,IF_15_Z,S_21_x,S_21_y,S_21_z= read_field(outputfolder+'21_21/')
IF_25_x,IF_25_y,IF_15_Z,S_25_x,S_25_y,S_25_z= read_field(outputfolder+'25_25/')
IF_51_x,IF_51_y,IF_51_Z,S_51_x,S_51_y,S_51_z= read_field(outputfolder+'51_51/')
IF_101_x,IF_101_y,IF_101_Z,S_101_x,S_101_y,S_101_z= read_field(outputfolder+'101_101/')
IF_201_x,IF_201_y,IF_201_Z,S_201_x,S_201_y,S_201_z= read_field(outputfolder+'201_201/')
```
