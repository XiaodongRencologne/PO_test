```python
from SOSATpopy2_filter import sosat
```

### 3-layer grooves, filter ARC

```python
dx_list = [180]
angle_list = [45]#[45,30,15,0,-15,-30,-45] 
freq_list = [90]
for freq in freq_list:
    for angle in angle_list:
        for item in dx_list:
            print(freq, angle,item)
            Model1 = sosat(freq,
                        feedpos = [item,0,0],
                        feedrot = [0,0,angle],
                        AR_file = 'AR_LIB/Ideal_AR.h5',
                        groupname= 'ideal_AR',
                        filter_file = None,
                        groupname2 = str(freq)+'GHz',
                        outputfolder = '../Data/output_polar_ideal_AR/'+str(freq)+'GHz/')
            factor1 = 2
            factor2 = 1
            if freq >90:
                factor1 = 3.5
                factor2 = 2.5
            Model1.run_po([[int(380*factor1 ),int(1880*factor2)],[int(240*factor1),int(2160*factor1)]],
                        [[int(370*factor1 ),int(2320*factor2)],[int(195*factor1),int(390*factor2)]],
                        [[int(340*factor1),int(2320*factor2)],[int(300*factor1),int(200*factor2)]],
                        [205,740],
                        [210,660],
                        [235,680],
                        [210,480],
                        sampling_type = 'less')

```

```python
Model1.plot_beam(cmap = 'inferno')
```
