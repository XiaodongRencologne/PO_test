```python
import numpy as np
import matplotlib.pyplot as plt
from Kirchhoffpy.dielectric_film import film

c = 299792458
```

## Film parameters
* 1. thickness: h = 2mm
* 2. n = 1.54
* 3. frequency range 80-90GHz

```python
h = 2
n = 1.54
n2 = 2
freq_list = np.linspace(10,100,101)
h_list = np.linspace(1,100,2001)
```

```python

Lambda_list = c*1000/(freq_list*10**9)
R, T = film(h,
            1,n,n2,
            0,
            Lambda_list,
            polarization = 'perpendicular')
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,5))
ax.plot(freq_list,10*np.log10(T))
ax.set_xlabel('freq (GHz)')
ax.set_ylabel('Transimission dB')
ax.grid('both')
```
