```python
import numpy as np
import matplotlib.pyplot as plt
from Kirchhoffpy.dielectric_film import film, Fresnel_coeffi, Fresnel_coeffi_power, critical_angle,brewster_angle
```

## Draw Fresnel_coeffi transmission coefficients


```python
n0 =1
n_si = 1.54
angle = np.linspace(0,89,501)
angle_in = angle * np.pi / 180  # angle in radians
angle_brewster = brewster_angle(n0, n_si)*180/np.pi
t_p_1,t_s_1,r_p_1,r_s_1 = Fresnel_coeffi(n0, n_si, angle_in)
T_p_1,T_s_1,R_p_1,R_s_1 = Fresnel_coeffi_power(n0, n_si, angle_in)

# silicon to aire
angle_critical = critical_angle(n_si, n0)*180/np.pi
angle_brewster2 = brewster_angle(n_si, n0)*180/np.pi
angle2 = np.linspace(0,89,501)
angle_in_2 = angle2*np.pi/180  # angle in radians
t_p_2,t_s_2,r_p_2,r_s_2 = Fresnel_coeffi(n_si, n0, angle_in_2)
T_p_2,T_s_2,R_p_2,R_s_2 = Fresnel_coeffi_power(n_si, n0, angle_in_2)

print(T_p_1[0], T_p_2[0],10*np.log10(T_p_1[0] * T_p_2[0]))
```

### 1. light from Air to Silicon
refractive index of Silicon is 3.36

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.plot(angle, np.abs(r_s_1), label='s-polar', color='blue',linestyle='--')
ax.plot(angle, np.abs(r_p_1), label='p-polar', color='red',linestyle='--')
ax.plot(angle2, np.abs(r_s_2), label='s-polar', color='blue',linestyle='-')
ax.plot(angle2, np.abs(r_p_2), label='p-polar', color='red',linestyle='-')
ax.set_xlim([0,90])
#ax.set_ylim([0,1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.plot(angle, np.abs(t_s_1), label='s-polar', color='blue',linestyle='--')
ax.plot(angle, np.abs(t_p_1), label='p-polar', color='red',linestyle='--')
ax.plot(angle2, np.abs(t_s_2), label='s-polar', color='blue',linestyle='-')
ax.plot(angle2, np.abs(t_p_2), label='p-polar', color='red',linestyle='-')
ax.set_xlim([0,90])
#ax.set_ylim([0,1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 9))
#ax.plot(angle, np.abs(R_s_1), label='s-polar R', color='blue',linestyle='--')
#ax.plot(angle, np.abs(R_p_1), label='p-polar R', color='red',linestyle='--')
ax.plot(angle, np.abs(T_s_1), label='s-polar, Transmission, Si to Air', color='blue',linestyle='-')
ax.plot(angle, np.abs(T_p_1), label='p-polar T, Si to Air', color='red',linestyle='-')
ax.plot(angle, np.abs(T_s_2), label='s-polar, T, Air to Si', color='blue',linestyle='--')
ax.plot(angle, np.abs(T_p_2), label='p-polar T, Air to Si', color='red',linestyle='--')
ax.axvline(x=angle_brewster, color='green', linestyle=':', label='Brewster angle')
ax.axvline(x=angle_critical, color='darkred', linestyle='-', label='Critical angle')
ax.set_xlim([0,90])
ax.set_ylim([0,1])
ax.set_title('Fresnel Coefficients',fontsize = 15)
plt.legend(fontsize= 13)
#plt.savefig('Fresnel_coefficients_silicon_air.png', dpi=300, bbox_inches='tight')
plt.savefig('Fresnel_coefficients_plastic_air.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 9))
ax.plot(angle2, np.abs(R_s_2), label='s-polar', color='blue',linestyle='--')
ax.plot(angle2, np.abs(R_p_2), label='p-polar', color='red',linestyle='--')
ax.plot(angle2, np.abs(T_s_2), label='s-polar', color='blue',linestyle='-')
ax.plot(angle2, np.abs(T_p_2), label='p-polar', color='red',linestyle='-')
ax.axvline(x=angle_brewster2, color='black', linestyle=':', label='Brewster angle')
ax.set_xlim([0,90])
ax.set_ylim([0,1])
```
