```python
import pyvista as pv
from Kirchhoffpy import lenspy
from Kirchhoffpy import Feedpy
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np
from Kirchhoffpy import coordinate,field_storage,Grids
from Kirchhoffpy.Vopy import CO,dotproduct,abs_v

from Kirchhoffpy.coordinate_operations import cartesian_to_spherical as cart2spher;
c=299792458
mu=4*np.pi*10**(-7);
epsilon=8.854187817*10**(-12);
Z0=np.sqrt(mu/epsilon,dtype = np.float64)
p = pv.Plotter()
srffolder = 'srf/'

from read_GRASP_output import read_grd
```

```python
freq = 90*10**9
Lambda = c*1000/freq
k = 2*np.pi/Lambda
print(Lambda,k)
```

```python

# 2. define coordinate systems
coord_ref = coordinate.coord_sys([0,0,0],[0,0,0],axes = 'xyz')

coord_L1 = coordinate.coord_sys([0,0,-960],[np.pi,0,0],axes = 'xyz',ref_coord = coord_ref)

coord_feed = coordinate.coord_sys([0,0,0],[np.pi,0,0],axes = 'xyz',ref_coord = coord_ref)

coord_sky = coordinate.coord_sys([0,0,0],[np.pi,0,0],axes = 'xyz',ref_coord = coord_ref)
# 3. define input Feedhorn
feed= Feedpy.GaussiBeam(-3,5,
                        k,
                        coord_feed,
                        polarization = 'x')
```

```python
x,y, G_co, G_cx, G_z = read_grd('Gaussian_far/plan3_grd.grd')
Max= np.abs(G_co).max()
NN = np.where(np.abs(G_co) == Max)
print(Max)
print((np.abs(G_co)**2).sum()*(x[1]-x[0])*(y[1]-y[0]))
#Max = 1
skip = 1
Max = G_co[NN]
G_co = G_co[::skip,::skip]#/Max
G_cx = G_cx[::skip,::skip]#/Max
G_z  = G_z[::skip,::skip]#/Max
```

```python
P1, n1 = Grids.plane(50,50,201,201)
P1.z[:] = 20
P2, n2 = Grids.plane(100,100,201,201)
P2.z[:] = 200
P3, n3 = Grids.plane(500,500,1001,1001)
P3.z[:] = 1000
P3.w =1
```

```python
E, H = feed.source(P3, k)
r,theta,phi=cart2spher(P3.x,P3.y,P3.z)
co,cx,crho=CO(theta,phi)
E_co = dotproduct(E,co)
E_cx = dotproduct(E,cx)
E_z = dotproduct(E,crho)
E_co = E.x
E_cx = E.y
E_z = E.z
Max= np.abs(E_co).max()

E_co = E_co.reshape(1001,-1)#/Max
E_cx = E_cx.reshape(1001,-1)#/Max
E_z = E_z.reshape(1001,-1)#/Max

err_co = np.abs(E_co) - np.abs(G_co)
err_cx = np.abs(E_cx) - np.abs(G_cx)
err_z = np.abs(E_z) -np.abs(G_z)
```
