```python
import pyvista as pv
from Kirchhoffpy import lenspy
from Kirchhoffpy import Feedpy,coordinate_operations,Grids
from Kirchhoffpy.coordinate_operations import Coord#,cart2spher

import matplotlib.pyplot as plt
import numpy as np
from Kirchhoffpy import LensPO

from read_GRASP_output import read_grd
c=299792458
```

```python
x,y, G_co, G_cx, G_z = read_grd('ellipGaussian/plan3_grd.grd')
Max= np.abs(G_co).max()
NN = np.where(np.abs(G_co) == Max)
print(Max)
print((np.abs(G_co)**2).sum()*(x[1]-x[0])*(y[1]-y[0]))
#Max = 1
Max = G_co[NN]
G_co = G_co/Max
G_cx = G_cx/Max
G_z  = G_z/Max
```

```python
freq = 90*10**9
Lambda = c*1000/freq
k = 2*np.pi/Lambda
```

```python
P1, n1 = Grids.plane(50,50,201,201)
P1.z[:] = 20
P2, n2 = Grids.plane(100,100,201,201)
P2.z[:] = 200
P3, n3 = Grids.plane(500,500,1001,1001)
P3.z[:] = 1000
```

```python
Ellip_beam = Feedpy.Elliptical_GaussianBeam([-3,-3],
                                            [5,10],
                                            k,[0,0,0],np.array([0,0,0]),
                                            polarization = 'x')
```

```python
E1,co,cx,c_z = Ellip_beam.beam(P3, n3)
Max = np.abs(co).max()
NN = np.where(np.abs(co) == Max)
print(Max)
Max = co[NN]
#Max = 1
print((np.abs(co)**2).sum()*(x[1]-x[0])*(y[1]-y[0]))
co =co.reshape(1001,-1)/Max
cx =cx.reshape(1001,-1)/Max 
c_z = c_z.reshape(1001,-1)/Max
err_co = np.abs(co) - np.abs(G_co)
err_cx = np.abs(cx) - np.abs(G_cx)
err_z = np.abs(c_z) -np.abs(G_z)
```
